Especifica la probabilidad de que una variable aleatoria sea menor o igual que un valor dado: $F(x) = P(X<=x)$

Definiremos una función que toma como entrada una matriz de datos de 1 Dimensión y luego devulve los valores de "x" y "y" de la función de distribución acumulada. 

Las ECDF se encuentran entre los gráficos más importantes en el análisis estadistico.

```python
#Función para calcular al función amumulada de una columna de datos.

def ecdf(data): 
    n = len(data)
    
    x = np.sort(data)
    y = np.arange(1, n+1)/n
    return x,y
```

Aplicando la función para obtener los valores de la función de distribución acumulada, al porcentaje de votos para Obahama.

**Varianza, desviación tipica**

La varianza es una medida de dispersión que representa la variabilidad de una serie de datos respecto a su media.Formalmente se calcula como la suma de los residuos al cuadrado divididos entre el total de observaciones.

También se puede calcular como la desviación típica al cuadrado.Dicho sea de paso, entendemos como residuo a la diferencia entre el valor de una variable en un momento y el valor medio de toda la variable.

Antes de ver la fórmula de la varianza, debemos decir que la varianza en estadística es muy importante.Ya que aunque se trata de una medida sencilla, puede aportar mucha información sobre una variable en concreto.

$s^{2}=\frac{{\displaystyle \sum_{i=1}^{n-1}\left(x-\bar{x}\right)^{2}}}{N}$

**¿Qué diferencia existe entre la varianza y la desviación típica?**

Una cuestión que se podría plantear, y con razón, sería la diferencia entre varianza y desviación típica. En realidad, vienen a medir lo mismo. 

La varianza es la desviación típica elevada al cuadrado. O al revés, la desviación típica es la raíz cuadrada de la varianza.

La desviación típica se hace para poder trabajar en las unidades de medida iniciales. Claro que, como es normal, uno puede preguntarse, ¿de qué sirve tener como concepto la varianza? Bien, aunque la interpretación del valor que arroja no nos da demasiada información, su cálculo es necesario para obtener el valor de otros parámetros.

$s=\sqrt{\frac{{\displaystyle \sum_{i=1}^{n-1}\left(x-\bar{x}\right)^{2}}}{N}}$

**Covarianza**

En probabilidad y estadística, la covarianza es un valor que indica el grado de variación conjunta de dos variables aleatorias respecto a sus medias.

Es el dato básico para determinar si existe una dependencia entre ambas variables y además es el dato necesario para estimar otros parámetros básicos, como el coeficiente de correlación lineal o la recta de regresión.

$Covarianza = 1/n \sum_{i=1}^{n} (x_{i}-\bar{x}) (y_{i}-\bar{y}) = 1/n \sum_{i=1}^{n} x_{i}y_{i} - \bar{x}\bar{y} $

 - Cuando entre las 2 variables hay una relación directa, la covarianza da un valor +.
 - Cuando entre las 2 variables hay una relación inversa, la covarianza da un valor -
 - Cuando entre las 2 variables no hay una relación, la covarianza da un valor en torno a 0

Una medida de cómo dos cantidades varían juntas

La covarianza se puede calcular utilizando la función Numpy $np.cov()$.

Por ejemplo, tenemos dos conjuntos de datos x y y, $np.cov(x, y)$ devuelve una matriz 2D donde las entradas [0,1]y [1,0] son las covarianzas.

La entrada [0,0] es la varianza de los datos en x, y la entrada [1,1] es la varianza de los datos en y.

Esta matriz de salida 2D se llama matriz de covarianza , ya que organiza la autovarianza y la covarianza

**Correlación**

En probabilidad y estadística, la correlación indica la fuerza y la dirección de una relación lineal y proporcionalidad entre dos variables estadísticas.

Se considera que dos variables cuantitativas están correlacionadas cuando los valores de una de ellas varían sistemáticamente con respecto a los valores homónimos de la otra: si tenemos dos variables (A y B) existe correlación entre ellas si al disminuir los valores de A lo hacen también los de B y viceversa.

La correlación entre dos variables no implica, por sí misma, ninguna relación de causalidad

Existen diversos coeficientes que miden el grado de correlación, adaptados a la naturaleza de los datos.

El más conocido es el coeficiente de correlación de Pearson (introducido en realidad por Francis Galton), que se obtiene dividiendo la covarianza de dos variables entre el producto de sus desviaciones estándar.

**Coeficiente de Correlación de Pearson**

$Pearson = Covarianza / (std(x) * std(y))$

= variabilidad debido a la codependencia /  variabilidad independiente 

**Interpretación**

El valor del índice de correlación varía en el intervalo [-1,1], indicando el signo el sentido de la relación:

- Si r = 1, existe una correlación positiva perfecta. El índice indica una dependencia total entre las dos variables denominada relación directa: cuando una de ellas aumenta, la otra también lo hace en proporción constante.
- Si 0 < r < 1, existe una correlación positiva.
- Si r = 0, no existe relación lineal. Pero esto no necesariamente implica que las variables son independientes: pueden existir todavía relaciones no lineales entre las dos variables.
- Si -1 < r < 0, existe una correlación negativa.
- Si r = -1, existe una correlación negativa perfecta. El índice indica una dependencia total entre las dos variables llamada relación inversa: cuando una de ellas aumenta, la otra disminuye en proporción constante.

**Ensayos Bernoulli**

Pensar en ensayos Bernoulli como un lanzamiento de moneda posiblemente sesgada.

Específicamente, cada lanzamiento de moneda tiene una probabilidad $p$ caer cara (éxito) y probabilidad $1 - p$ caer cruz (falla).

Veamos una función para realizar $n$ pruebas de Bernoulli "perform_bernoulli_trials(n, p)", que devuelve el número de éxitos de las $n$ pruebas Bernoulli, cada uno de los cuales tiene probabilidad $p$ de éxito.

Para realizar cada prueba Bernoulli, usando la "np.random.random()" función, que devuelve un número aleatorio entre cero y uno.

```python
def perform_bernoulli_trials(n, p):
    """Realizar n ensayos de Bernoulli con probabilidad de éxito p y devolver el número de éxitos """
    n_success = 0

    # Realizar n ensayos
    for i in range(n):
        # número aleatorio entre cero y uno
        random_number = np.random.random()

        # Si es menor que p, es un éxito, se suma uno a n_success
        if random_number < p:
            n_success += 1

    return n_success
```

**número de éxitos de las  𝑛  pruebas Bernoulli, cada uno de los cuales tiene probabilidad  𝑝  de éxito.**

```python
#10 prueba bernuolli con p = .5 (probabilidad de exito)
perform_bernoulli_trials(10,p=.5)
```

**¿Cuántos valores predeterminados podríamos esperar?**

Digamos que un banco hizo 100 préstamos hipotecarios.

Es posible que entre 0 y 100 de los préstamos sean incumplidos.

Nos gustaría saber la probabilidad de obtener un número determinado de valores predeterminados, dado que la probabilidad de un incumplimiento es p = 0.05.

Para investigar esto, haremos una simulación.

Realizando 100 pruebas de Bernoulli utilizando la "perform_bernoulli_trials()" función escrita arriba, registrando cuántos valores predeterminados tenemos.

Aquí, un éxito es un defecto. (Recordar que la palabra "éxito" solo significa que el ensayo de Bernoulli se evalúa True, es decir, ¿el incumplimiento del destinatario del préstamo?)

Lo haremos con 100 ensayos de Bernoulli.

Una y otra vez hasta que lo hayamos probado 1000 veces.

Luego, trazeremos un histograma que describe la probabilidad del número de valores predeterminados.

```python
plt.rcParams['figure.figsize'] = [25, 6]

#semilla
np.random.seed(42)

# Inicializar el número de valores predeterminados: n_defaults
n_defaults = np.empty (1000)

# Calcular el número de valores predeterminados
for i in range(1000):
    n_defaults [i] = perform_bernoulli_trials(100, 0.05)

# histograma con el número predeterminado de bins
_ = plt.hist(n_defaults, normed = True)
_ = plt.xlabel('conteo de número de incumplimientos de 100 préstamos')
_ = plt.ylabel('probabilidad')
_ = plt.title('Simulación de 100 prestamos donde p = .05 (incumplimiento de pago)')

plt.annotate('podriamos decir que entre 4 y 8 incumplimiento por cada 100 prestamos',
              xy = (6, .10),
              xytext = (10, .08), 
              #  Shrink the arrow to avoid occlusion
              arrowprops = {'facecolor':'red', 'width': 5, 'shrink': 1})

plt.show ()

#En realidad, esta no es una forma óptima de trazar un histograma cuando se sabe que los resultados son enteros.
#mas adelante veremos una mejor opción para visualizar.
```

**¿El banco fallará?**

Trazemos el número de valores predeterminados que se obtuvo  en el espacio de n_defaults, como un ECDF Funciones de distribucion acumulativa empirica.

Si las tasas de interés son tales que el banco perderá dinero si se dejan de pagar 10 o más de sus préstamos.

**¿cuál es la probabilidad de que el banco pierda dinero?**

```python
# Funcion de distribucion acumulativa se definio en la primera parte.
x, y = ecdf(n_defaults)

#grafica
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('numero de incumplimientos en 100 prestamos')
_ = plt.ylabel('Función de distribución acumulativa')
_ = plt.title('Distribución Acumulada de prestamos incumplidos simulando Bernoulli 100 veces')

plt.annotate('Mayor densidad a los 8 prestamos o menos',
              xy = (8, .8),
              xytext = (9, .9), 
              #  Shrink the arrow to avoid occlusion
              arrowprops = {'facecolor':'red', 'width': 5, 'shrink': 1})

plt.show()

# Calculo de la cantidad de simulaciones de 100 préstamos con 10 o más valores predeterminados: n_lose_money
n_lose_money = np.sum(n_defaults >= 10)

# Calcular e imprimir la probabilidad de perder dinero
print('Probabilidad de perder dinero =', n_lose_money / len(n_defaults))
```

**Distribución Binomial**

En estadística, la distribución binomial es una distribución de probabilidad discreta que cuenta el número de éxitos en una secuencia de n ensayos de Bernoulli independientes entre sí, con una probabilidad fija **p** de ocurrencia del éxito entre los ensayos.

Un experimento de Bernoulli se caracteriza por ser dicotómico, esto es, solo dos resultados son posibles.

A uno de estos se denomina **«éxito»** y tiene una probabilidad de ocurrencia **p** y al otro, **«fracaso»**, con una probabilidad **q = 1 - p**.

En la distribución binomial el anterior experimento se repite n veces, de forma independiente, y se trata de calcular la probabilidad de un determinado número de éxitos.

- **Para n = 1, la binomial se convierte, de hecho, en una distribución de Bernoulli.**

Para representar que una variable aleatoria X sigue una distribución binomial de parámetros **n** y **p**, se escribe:

${\displaystyle X\sim B(n,p)\,}$

La distribución binomial es la base del test binomial de significación estadística.


**Calcularemos la distribución acumulada para el número de incumplimientos que esperaríamos para 100 préstamos**

Como en la parte de arriba, pero en lugar de simular todos los ensayos de Bernoulli, utilizaremos **np.random.binomial()**.

Esto es idéntico al cálculo de arriba usando la función **perform_bernoulli_trials()** función personalizada, pero mucho más eficiente computacionalmente.

Dada esta eficiencia adicional, tomaremos 10,000 muestras en lugar de 1000.

Después de tomar las muestras, graficaremos el ECDF como la última vez.

Este ECDF que está trazando es el de la distribución Binomial.


```python
# Tomando 10,000 muestras de la distribución binomial: n_defaults
n_defaults = np.random.binomial(n=100, p=0.05, size=10000)

# Computando ecdf (funcion de distribución acumulativa empirica)
x, y = ecdf(n_defaults)

# Grafica
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('numero de incumplimientos en 100 prestamos')
_ = plt.ylabel('Función de distribución acumulativa')
_ = plt.title('Distribución Acumulada de prestamos incumplidos simulados con una Binomial(100, p = 0.05)')

plt.annotate('Simulación usando np.random.binomial(), similar a cuando generamos los eventos Bernoulli',
              xy = (8, .8),
              xytext = (9, .9), 
              #  Shrink the arrow to avoid occlusion
              arrowprops = {'facecolor':'red', 'width': 5, 'shrink': 1})

plt.show()
```

**Función de masa de probabilidad (PMF)**

Conjunto de probabilidades de resultados discretos.

Es la dristribución de probabilidad de los posibles resultados en un experimento.

Ejemplo: X = númuero de "soles" en 3 volados de una moneda justa.

Espacio Muestral =  [SSS, SSA, SAS, SAA, ASS, AAS, ASA, AAA]

Con esto podemos determinar:

- $P(X = 0) = 1/8$ (AAA) cero soles

- $P(X = 1) = 3/8$ (SAA,AAS,ASA) un sol

**Poisson**

El momento del próximo evento es completamente independiente de cuando ocurrió el evento anterior. (eje; nacimientos en un hospital, éxitos de un sitio web, )

    El número r de llegadas de un proceso de Poisson en un intervalo de tiempo dado con una tasa promedio de λ llegadas por intervalo está distribuido por Poisson.
    
    El número r de visitas a un sitio web en una hora con una tasa de visitas promedio de 6 visitas por hora se distribuye por Poisson.
    
Límite de la distribución binomial para baja probabilidad de éxito y gran número de ensayos.
Es decir, para eventos raros.

##### Relación entre distribución Binomial y Poisson.

La distribución Poisson es un límite de la distribución binomial para eventos raros

Esto tiene sentido si piensa en las historias. Digamos que hacemos una prueba de Bernoulli cada minuto durante una hora, cada una con una probabilidad de éxito de 0,1.

Haríamos 60 pruebas, y el número de éxitos se distribuye Binomialmente, y esperaríamos obtener alrededor de 6 éxitos.

Donde obtenemos un promedio de 6 visitas a un sitio web por hora.

Entonces, la distribución de Poisson con una tasa de llegada igual a np se aproxima a una distribución Binomial para n ensayos de Bernoulli con probabilidad p de éxito (con n grande y p pequeño).

Es importante destacar que, a menudo, es más sencillo trabajar con la distribución de Poisson porque solo tiene un parámetro en lugar de dos para la distribución Binomial.

##### Exploremos estas dos distribuciones computacionalmente.

Calculando la media y la desviación estándar de las muestras de una distribución de Poisson con una tasa de llegada de 10.

Luego, la media y la desviación estándar de las muestras de una distribución Binomial con los parámetros n y p tales que np = 10.

```python
# 10,000 muestras de la distribución de Poisson: samples_poisson
samples_poisson = np.random.poisson(10, size=10000)

# Media y desviación de Poisson
print('Poisson:     ', np.mean(samples_poisson),
                       np.std(samples_poisson))

# Especificando los valores de n y p considerando una Binomial: n, p
n = [20, 100, 1000]
p = [0.5, 0.1, 0.01]

# Dibujar 10,000 muestras para cada par n, p: samples_binomial
for i in range(3):
    samples_binomial = np.random.binomial(n[i], p[i], size=10000)

    # Resultados
    print('n =', n[i], 'Binom:', np.mean(samples_binomial),
                                 np.std(samples_binomial))
```

Las medias son prácticamente las mismas, lo que puede demostrarse con un trabajo de lápiz y papel.

La desviación estándar de la distribución binomial se acerca cada vez más a la de la distribución de Poisson a medida que la probabilidad p disminuye cada vez más.

##### ¿Cuántos juegos sin hits en una temporada? En béisbol

Un juego sin hits es un juego en el que un lanzador no permite que el otro equipo consiga un hit.

Este es un evento raro, y desde el comienzo de la llamada era moderna del béisbol (a partir de 1901), solo ha habido 251 de ellos hasta la temporada 2015 en más de 200,000 juegos.

El ECDF del número de juegos sin hits en una temporada se muestra abajo. ¿Qué distribución de probabilidad sería apropiada para describir la cantidad de juegos sin hits que esperaríamos en una temporada determinada?


```python
n_nohitters = np.random.poisson(251/115, size=10000)

x, y = ecdf(n_nohitters)

_ = plt.plot(x, y, marker='.', linestyle='none')

_ = plt.xlabel('numero de no hitters en la temporada')

_ = plt.ylabel('ECDF')

plt.show()
```

¿2015 fue anómalo? 1990 y 2015 presentaron la mayor cantidad de juegos sin hits de cualquier temporada de béisbol (hubo siete).

Dado que hay un promedio de 251/115 partidos sin hits por temporada, ¿cuál es la probabilidad de tener siete o más en una temporada?

```python
# Extraiga 10,000 muestras de la distribución de Poisson: n_nohitters
n_nohitters = np.random.poisson(251/115, size=10000)

# Calcular el número de muestras que son siete o más: n_large
n_large = np.sum(n_nohitters >= 7)

# Calcule la probabilidad de obtener siete o más: p_large
p_large = n_large / 10000

print('Probabilidad de siete o más juegos sin hits:', p_large)
```

El resultado es de aproximadamente 0,007.

Esto significa que no es tan improbable ver una temporada sin hits de 7 o más en un siglo. Hemos visto dos en un siglo y medio, por lo que no es descabellado.
