## Parte 3 Reorganizacion de los datos

Pivoteando una sola variable, melt, stack y unstack, pivot table

```python
gapminder = pd.read_csv('https://assets.datacamp.com/production/repositories/502/datasets/09378cc53faec573bcb802dce03b01318108a880/gapminder_tidy.csv')
gapminder.head()
```

```python
#Se cuentan 50 paises
print(pd.unique(gapminder.Country.value_counts()))
#49 años
print(gapminder.Year.max() - gapminder.Year.min())
```

Nos gustaria tener un resumen de la pertenencia de los paises por region, al pivotear una variable no podemos tener datos duplicados.

Por lo que solo trabajaremos con un año 1964

```python
gapminder = gapminder.set_index(['Country','Year']) 
gapminder_copy = gapminder.loc[(slice(None),1964),:] #Filtramos solo los valores del año 1964 para no tener datos duplicados
gapminder_copy = gapminder_copy.reset_index() #.pivot no permitira trabajar con los index
```

```python
gapminder_copy.pivot(index = 'region', columns='Country', values = 'gdp')
```

```python
#Si no especificamos la columna a pivotear todas las de mas columnas seran pivoteadas
gapminder_copy.pivot(index = 'region', columns = 'Country')
#test.unstack(level = 'region')
```

#### Stack y unstack

```python
#Desapiland los datos
gapminder.unstack(level = 'Country').head()
```

```python
#Apilando los datos columnas en filas
gapminder.unstack(level = 'Country').stack(level='Country').head()
```

```python
#Cambia los niveles de index rellena los valores.
gapminder.unstack(level = 'Country').stack(level='Country').swaplevel(0,1).head()
```

#### Melt

Se especifica la columna que quedara fija, mientras las demas columnas se almacenaran en una sola columna junto con sus valores.

```python
gapminder = gapminder.reset_index()
```

```python
pd.melt(gapminder, id_vars= 'Country', value_name='valores').tail()
```

```python
#Especificamos 3 columnas y el resto se coloca en variable y value
pd.melt(gapminder, id_vars= ['region','Country','Year']).head()
```

```python
#A veces, todo lo que necesitamos son algunos pares clave-valor, y el contexto no importa. 
#Si dicho contexto está en el índice, se puede obtener fácilmente lo que se dese.
pd.melt(gapminder.set_index(['region', 'Country','Year']), col_level=0).head()
```

## Parte 4 Agrupacion de Datos

```python
gapminder_copy = gapminder.copy()
gapminder_copy = gapminder_copy[gapminder.Year == 2010]
gapminder_copy.groupby(['region','Country']).mean().sort_index().head(10)
```

```python
#Podemos colocar un pandas Series en groupby
regions = pd.Series(gapminder.region.unique())
gapminder.groupby(regions).mean()
```

#### Groupby con agregaciones personalizadas y diccionarios

Al crear nuestras propias funciones podemos aplicarles groupby con el metodo .agg, tambien podemos especificar funcion aplicar a cada columna

```python
def max_min(Series):
    return Series.max() - Series.min()

gapminder.groupby(['region','Country']).agg(max_min).head()
```

```python
#Colocamos un diccionarion con las funciones a aplicar para cada columna
gapminder.groupby(['region','Country']).agg({'fertility':'count','gdp':max_min,'life':'mean'}).head()
```

Transformaciones en groupby

```python
standardized = gapminder_copy.groupby('region')['fertility','life'].transform(zscore)
outliers = (standardized['life'] < -3) | (standardized['fertility'] > 3)
gapminder_copy.loc[outliers]
```

Relleno de daltos faltantes, muchos paquetes estadisticos y de aprendizaje automatico no pueden determinar la mejor accion a tomar cuando se encuentran datos faltantes. La practica usual es .dropna(), veamos como rellenar datos faltantes con groupby

```python
#Para este ejemplo generaremos datos faltantes en la base que luego remplazaremos con la funcion imput_mean()
too_close = gapminder_copy['life'] > 80
#Asigna NaN, donde se cumpla la condicion
gapminder_copy.loc[too_close,'life'] = np.nan
print(gapminder_copy.notna().all())

#Remplazamos valores faltantes
def impute_mean(x):
    return 10

gapminder_copy[gapminder_copy.life.isna()].head()
```

Creamos las agrupaciones

```python
Grupos = gapminder.groupby(['Country','Year'])
Grupos.life = Grupos.life.transform(impute_mean)
```

Con las agrupaciones remplazamos los NaN en la base original

```python
gapminder_copy.life = Grupos.life.transform(impute_mean)
```

#### Transformacion .apply()

```python
regional = gapminder_copy.groupby('region')
```

```python
def disparity(gr):
    s = gr['gdp'].max() - gr['gdp'].min()
    z = (gr['gdp'] - gr['gdp'].mean())/gr['gdp'].std()
    return pd.DataFrame({'z(gdp)':z , 'regional spread(gdp)':s})
```

```python
regional.apply(disparity).head()
```
