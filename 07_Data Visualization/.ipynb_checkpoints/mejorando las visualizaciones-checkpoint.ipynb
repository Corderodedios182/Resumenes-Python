```python
import pandas as pd

marketing = pd.read_csv("/home/carlos/Documentos/Resumenes-Python/databases to have fun/Base master Roas - semanal normal.csv", 
                        parse_dates = ['fecha_inicio_plan','fecha_fin_plan','inicio_reporte','fin_reporte'])

marketing = marketing.fillna(0)

marketing['inversión_total'] = marketing['inversión_total'].astype(int)
marketing['ctr'] = (marketing['clics'] / marketing['impresiones']).round(3)

marketing.keys()
```

```python
marketing.groupby(['cliente_nomenclatura']).size()
```

## Destacando los datos

- Estilos
- Marcado de puntos
- Anotaciones

**Estilo, pelata de colores, subtitulos en nuestra gŕafica**

```python
import seaborn as sns
import matplotlib.pyplot as plt 

#Podemos jugar con el estilo de gráficas
plt.style.use("ggplot")

ths = marketing[marketing['cliente_nomenclatura'] == 'ths']

#Parametro pelette, permite cambiar la pelata de colores
g = sns.catplot(x = 'plataforma',
                data = ths, 
                kind = "count",
                col = "Año-Mes",
                palette = "husl")

#Rotación de ejes
g.set_xticklabels(rotation=90)

plt.show()
```

**Reglas para colorear ciertos puntos de interés, colocación de anotaciones**

```python
#Podemos cuidar el tamaño de nuestras gráficas en los cuadernos jupyter
plt.rcParams['figure.figsize'] = [10, 5]

#Filtro de datos
ths = ths.groupby(['Año-Mes','plataforma','campaña_nomenclatura.1'],as_index = False).sum()[['Año-Mes','plataforma','campaña_nomenclatura.1','inversión_total','dinero_gastado','impresiones','clics']]
ths['ctr'] = (ths['clics'] / ths['impresiones']).round(3)

#Creamos una regla para colorear ciertos puntos
tmp = ['purple' if (inversión_total > 10000) & (ctr > .03) else 'navy' for inversión_total, ctr in zip(ths.inversión_total, ths.ctr) ]

#scatterplot
g = sns.regplot(x='clics',
                y='impresiones',
                data = ths,
                fit_reg = False,
                scatter_kws = {'facecolors': tmp, 'alpha': 0.7}).set_title("Es bueno desviar la atención a ciertos puntos de la gráfica, para esto crearemos una regla pintando los puntos con ctr mayor a 3%.")

#Colocación de anotaciones
plt.text(60000, 1000000,
         'Campañas con altos clics e impresiones', 
          color = 'black',
          # Set the font to large
          fontdict = {'ha': 'left', 'size': 'large'})

plt.show()
```

**Desviando la atención a un solo punto**

```python
#Filtro de datos
tmp = marketing.loc[((marketing['clics'] < 100 ) &
                    (marketing['impresiones'] < 300000)),:]

#Desviaremos la atención en el valor máximo
max_ctr = tmp.ctr.max()

#Creamos una regla para marcar el valor maximo
tmp['color_ctr'] = ['Mayor ctr' if ctr == max_ctr else 'Otros' for ctr in tmp['ctr']]


#scatterplot
g = sns.scatterplot(x='clics',
                    y='impresiones',
                    data = tmp,
                    hue = 'color_ctr',
                    palette = 'Set2').set_title("Coloreando la campaña con un mayor ctr")

#especificamos donde comienza la indicación y la etiqueta de la antoación
plt.annotate('Campaña con el mayor ctr 1.67',
              xy = (tmp.query("ctr == 1.167")["clics"], tmp.query("ctr == 1.167")["impresiones"]),
              xytext = (10, 50000), 
              #  Shrink the arrow to avoid occlusion
              arrowprops = {'facecolor':'gray', 'width': 3, 'shrink': 0.03},
              backgroundcolor = 'black')

plt.show()
```

**Comparando Resultados Mes a Mes**

```python
sns.kdeplot(marketing[marketing['Año-Mes'] == 2001].ctr, 
            # Shade under kde and add a helpful label
            shade = True,
            label = 'Enero')

sns.kdeplot(marketing[marketing['Año-Mes'] == 2002].ctr, 
            # Shade under kde and add a helpful label
            shade = True,
            label = 'Febrero')

sns.kdeplot(marketing[marketing['Año-Mes'] == 2003].ctr, 
            # Shade under kde and add a helpful label
            shade = True,
            label = 'Marzo').set_title("Ajuste y grafica de una estimación de densidad de grano bivariada, ctr")

plt.show()
```

Una forma de mejorar los KDE es con la adición de un diagrama de alfombra . Los diagramas de alfombra son pequeños guiones dibujados debajo de la densidad que muestran con precisión dónde cae cada punto de datos. Agregar un diagrama de alfombra es particularmente útil cuando no tienes una tonelada de datos.

Con pequeñas cantidades de datos, a menudo tiene lagunas en su soporte sin datos, y puede ser difícil saber si una línea KDE distinta de cero significa que los datos estaban presentes o se deben a un núcleo ancho. Un diagrama de alfombra ayuda a abordar esto.

```python
sns.distplot(marketing[marketing['Año-Mes'] == 2001].ctr,
             label = "Enero",
             hist = False,
             color = "gray",
             rug = True)

sns.distplot(marketing[marketing['Año-Mes'] == 2002].ctr,
             label = "Febrero",
             hist = False,
             color = "steelblue",
             rug = True).set_title("Una forma de mejorar los KDE es con la adición de un diagrama de alfombra")


plt.show()
```

**Diagrama de abejas**

Recomendado para ver la acumulación de los datos.

```python
# Graficas de abeja, útil para ver la acumulación de los datos
sns.swarmplot(y = "plataforma",
              x = "ctr", 
              data = marketing[marketing['Año-Mes'] == 2001], 
              # Decrease the size of the points to avoid crowding 
              size = 3)

# Un título descriptivo
plt.title('Distribución del ctr por plataforma en el Mes de Enero')

plt.show()
```

## Jugando con los colores individuales y las paletas de colores.

- Colocar paletas de colores
- Uso de paletas de colores personalizadas
- Color categorias ordinales

**Elegir la variable correcta para codificar con color**

```python
plt.style.use("dark_background")

#Filtro
tmp = marketing.groupby(['Año-Mes','cliente_nomenclatura','plataforma'], as_index = False).sum()[['inversión_total','dinero_gastado','Año-Mes','cliente_nomenclatura','plataforma']]

tmp = tmp.query("inversión_total > 50000")

#--graficas--##

plataformas = ['Facebook', 'Google Search', 'Programmatic']

# filtro de plataformas
plataformas_maxes = tmp[tmp.plataforma.isin(plataformas)]

# grafica
g = sns.catplot(x = 'Año-Mes', 
                hue = 'cliente_nomenclatura',
                y = 'inversión_total',
                col = 'plataforma',
                data = plataformas_maxes,
                # Cambio de una peleta de colores 
                palette = 'Set2',
                sharey = False,
                kind = 'bar')

# subtitulos
plt.subplots_adjust(top=0.8)
g.fig.suptitle('Cuentas con una inversión mayor a $50,000')

plt.show()
```

**Tratar con demasiadas catogorías y partiendo los datos con FacetGrid**

```python
plt.style.use("seaborn")

g = sns.FacetGrid(data = tmp,
                  col = 'cliente_nomenclatura',
                  col_wrap = 3)

# Map sns.scatterplot to create separate city scatter plots
g.map(sns.scatterplot,
      'inversión_total',
      'dinero_gastado',
       alpha = 0.8,
       color = "#9b59b6")

plt.subplots_adjust(top = .9)
g.fig.suptitle('Inversión total y dinero por cliente')

plt.show()
```

**Arreglando los gráficos de barras de Seaborn**

```python
import numpy as np

sns.barplot(y = 'dinero_gastado', x = 'cliente_nomenclatura', 
            estimator = np.mean,
            ci = False,
            data = marketing,
            # Agrega un border a las barras
            edgecolor = 'cadetblue').set_title("El parametro edgecolor nos ayuda a darle un diseño agradable de colores")
```

**Paleta continua personalizada**

```python
# Filtro de datos
tmp = marketing.groupby(['Año-Mes','cliente_nomenclatura','plataforma'], as_index = False).agg({'inversión_total':'sum','dinero_gastado':'sum'})

# Definición de paleta personalizada
color_palette = sns.light_palette('navy',
                                  as_cmap = True)

# Coloreado de puntos pode medio de paletas
sns.scatterplot(x = 'dinero_gastado',
                y = 'inversión_total',
                hue = 'inversión_total', 
                data = tmp,
                palette = color_palette).set_title('Color de inversión')
plt.show()
```

**Ajustando la paleta de colores de acuerdo al contexto de datos paletas divergentes**

```python
#Filtro
tmp = marketing.groupby(['plataforma','inicio_reporte']).size().unstack()

# Define a custom palette
color_palette = sns.diverging_palette(250, 0,
                                      center = 'dark',
                                      as_cmap = True)

# Pass palette to plot and set axis ranges
sns.heatmap(tmp,
            cmap = color_palette,
            center = 30,
            vmin = 0,
            vmax = 60).set_title('Podemos especificar un rango de paleta de colores en nuestros heatmap')

plt.yticks(rotation = 0)
plt.show()
```

**Usanto paleta personalizada para categorías**

Tratando con demasiadras categorias y escojer los datos necesarios para colorear, debemos tener la facilidad de filtrar y categorizar nuestros datos.

```python
plt.style.use("dark_background")
plt.rcParams['figure.figsize'] = [25, 5]

# Color lines by the city and use custom ColorBrewer palette
g = sns.lineplot(x = "inicio_reporte",
                 y = "inversión_total",
                 hue = "plataforma",
                 palette = "Set2",
                 linewidth = 3,
                 data = marketing)


g.set_title("Inversión semanal")

plt.setp(g.get_xticklabels(), rotation=45)

plt.show()
```
