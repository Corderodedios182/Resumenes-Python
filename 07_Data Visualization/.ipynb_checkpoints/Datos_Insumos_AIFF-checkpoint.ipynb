```python
#PO_NBR : codigo orden de compra
#ORDER_DATE :
#WHSE_NBR : Codigo del Cedis
#ORIGINAL_CANCEL_DT : fecha ideal de entrega 
#VENDOR_NBR_DEPT :   
#PO_TYPE : tipo de orden de compra 
#VENDOR_NAME : nombre del provedor
#BOOKING_PO_NBR : 
#EVENT_DESC : descripcion del evento

purchase_order = pd.read_sql_query("""SELECT TOP 10
                                        PO_NBR, ORDER_DATE, WHSE_NBR, ORIGINAL_CANCEL_DT, VENDOR_NBR_DEPT,
                                        PO_TYPE, VENDOR_NAME, BOOKING_PO_NBR, EVENT_DESC
                                      FROM MX_CF_VM.PURCHASE_ORDER""", connection)

#Numero de columnas
purchase_order.head()
```

```python
#Dependiendo del tipo de articulo es el tipo de entrega.
#ejemplo: 20 = SST, 33 cross dock
#cross dock == xdock == cruce de anden: en menos de 24 horas se debe entregar
#staple stock == SSK: almacenar la mercancia en unos rack

pd.read_sql_query("""SELECT DISTINCT PO_TYPE
                     FROM MX_CF_VM.PURCHASE_ORDER
                      ORDER BY PO_TYPE DESC""", connection)
```

```python
pd.read_sql_query("""SELECT DISTINCT EVENT_DESC 
                     FROM MX_CF_VM.PURCHASE_ORDER""", connection)
```
