```python
import pandas as pd

marketing = pd.read_csv("/home/carlos/Documentos/Resumenes-Python/databases to have fun/Base master Roas - semanal normal.csv", 
                        parse_dates = ['fecha_inicio_plan','fecha_fin_plan','inicio_reporte','fin_reporte'], 
                        index_col = 'inicio_reporte')
   
marketing.head(2)
```

La visualización de datos en diagramas y figuras expone los patrones subyacentes en los datos y proporciona información.

Las buenas visualizaciones también ayudan a comunicar los datos a otros, y son útiles para los analistas de datos y otros consumidores de los datos.

Comenzando con ax.plot con 2 métricas en el eje x y y

**Gráficas de líneas**

```python
tmp = marketing.groupby(['cliente_nomenclatura'], as_index = False).sum()

tmp = tmp.loc[:,['cliente_nomenclatura','inversión_total','dinero_gastado']]

##--gráficas--##
import matplotlib.pyplot as plt

#Diferentes estilos para nuestras gráficas
plt.style.use('ggplot')

#Podemos cuidar el tamaño de nuestras gráficas en los cuadernos jupyter
plt.rcParams['figure.figsize'] = [20, 6]

#Primero definimos fig y ax
fig, ax = plt.subplots()

#Vamos agragendo componentes a ax
ax.plot(tmp.cliente_nomenclatura, tmp.inversión_total, color = "b", label = "a")
ax.plot(tmp.cliente_nomenclatura, tmp.dinero_gastado,marker="x",color="g", linestyle="--", label = "aaa")

#leyendas en los ejes
ax.set_xlabel("clientes")
ax.set_ylabel("$ dinero")

ax.set_title("inversión vs dinero gastado")

# Despacha la gráfica
plt.show()
```

**Podemos generar una matriz de gráficas con subplot.**

```python
tmp = marketing.groupby(['cliente_nomenclatura'], as_index = False).sum()

tmp = tmp.loc[:,['cliente_nomenclatura','inversión_total','dinero_gastado','impresiones','clics']]

##--gráficas--##

#subplots genera una matriz nxn y sharey = True funciona para que nuestras gráficas no se empalmen
fig, axs = plt.subplots(2,2, sharey = True)

#Cada función para hacer gráficas tiene parametros de diseño ejemplo marker = "o", color = "g", linestyle = "--"
axs[0,0].plot(tmp.cliente_nomenclatura, tmp.inversión_total, marker = "o")

axs[0,1].scatter(tmp.cliente_nomenclatura, tmp.inversión_total,marker = "o", color = "g", cmap = "viridis")

axs[1,0].plot(tmp.cliente_nomenclatura, tmp.inversión_total,linestyle = "--")

axs[1,1].bar(tmp.cliente_nomenclatura, tmp.inversión_total, color = "b")

for ax in axs.flat:
    ax.set_xticklabels(tmp.cliente_nomenclatura,rotation=90)
    ax.set(xlabel='clientes', ylabel='$ dinero')
    
for ax in axs.flat:
    ax.label_outer()
```

**Podemos colocar anotaciones relevantes en nuestros gráficos**

```python
tmp = marketing.groupby(['inicio_reporte']).sum()

tmp = tmp.loc[:,['inversión_total','dinero_gastado','impresiones','clics']]

#graficas
fig, ax = plt.subplots()

ax.plot(tmp.index, tmp.inversión_total, marker = "o", color = "blue")

ax.plot(tmp.index, tmp.dinero_gastado, linestyle = "--", color = "green")

ax.set_xticklabels(tmp.index,rotation=45)

ax.set_xlabel('tiempo')
ax.set_ylabel('$ dinero')

ax.set_title("dinero gastado por semana")

plt.show()
```

**Para no repetir el código de una función es útil definir una función que nos ayude con visualización.**

```python
# Define a function called plot_timeseries
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  # Plot the inputs x,y in the provided color
  axes.plot(x, y, color=color)

  # Set the x-axis label
  axes.set_xlabel(xlabel)

  # Set the y-axis label
  axes.set_ylabel(ylabel, color=color)

  # Set the colors tick params for y-axis
  axes.tick_params('y', colors=color)
```

```python
fig, ax = plt.subplots()

plot_timeseries(ax, tmp.index, tmp.impresiones, 'green', "inicio_reporte", "impresiones")

#ayuda a que la gráfica sea de 2 ejes
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, tmp.index, tmp.clics, 'red', "inicio_reporte", "clics")

ax.set_title("impresiones y clics por semana")

ax.set_xticklabels(tmp.index,rotation=45)

plt.show()
```

**Podemos ir agrupando los datos de la siguiente manera.**

```python
tmp = marketing.groupby(['cliente_nomenclatura','plataforma'])['plataforma'].count().unstack().reset_index().fillna(0)

#graficas
fig, ax = plt.subplots()

ax.bar(tmp.cliente_nomenclatura, tmp['Programmatic'], label="Programmatic")

ax.bar(tmp.cliente_nomenclatura, tmp['Google Search'], bottom=tmp['Programmatic'], label="Google Search")

ax.bar(tmp.cliente_nomenclatura, tmp['Facebook'], bottom=tmp['Programmatic'] + tmp['Google Search'], label="Facebook")

ax.set_title("conteo de campañas en plataformas por cliente")

plt.show()
```

**Histogramas**

Los histogramas nos permiten ver las distribuciones de los datos en diferentes grupos en nuestros datos.

```python
fig, ax = plt.subplots()

# Plot a histogram of "Weight" for mens_rowing
ax.hist(marketing.conversiones, histtype='step', label="conversiones", bins=50)

# Compare to histogram of "Weight" for mens_gymnastics
ax.hist(marketing[marketing.clics < 200]['clics'], histtype='step', label="clics", bins=50)

ax.set_xlabel("conteo")
ax.set_ylabel("# of observations")

ax.set_title("Histograma de conversiones y clics")

# Add the legend and show the Figure
ax.legend()
plt.show()
```

**Agregar barras de error a un gráfico de barras**

Las técnicas de trazado estadístico agregan información cuantitativa para realizar comparaciones en la visualización

```python
fig, ax = plt.subplots()

#Agregamos una linea para medir la desviación estandar, diferencia entre la media de nuestros datos.
ax.bar("clics", marketing["clics"].mean(), yerr=marketing["clics"].std())

ax.set_ylabel("n°")

ax.set_title("Podemos agregar un línea para marcar la desviación estándar")

plt.show()
```

**Colocar desviación estándar sobre cada punto**

```python
tmp = marketing.groupby(['fin_reporte']).describe()['conversiones']

#--graficas--#

fig, ax = plt.subplots()

#Agregamos una línea que indica la desviación estandar para cada punto.
ax.errorbar(tmp.index, tmp['mean'], yerr = tmp['std'])

ax.set(xlabel='fin del reporte', ylabel='conversiones promedio')
ax.set_title("Con errorbar a cada punto podemos indicar su desviación")

ax.set_xticklabels(tmp.index,rotation=45)

plt.show()
```

**Creación de boxplots**

```python
fig, ax = plt.subplots()

ax.boxplot(tmp.dinero_gastado)

plt.show()
```

```python
import numpy as np

plataformas = np.array(marketing["plataforma"].unique())
```

**Automatización de figuras**

```python
fig, ax = plt.subplots()

# Loop over the different sports branches
for sport in plataformas:
  # Extract the rows only for this sport
  sport_df = marketing[marketing["plataforma"] == sport]
  # Add a bar for the "Weight" mean with std y error bar
  ax.bar(sport, sport_df["inversión_total"].mean(), yerr=sport_df["inversión_total"].std())

ax.set_ylabel("inversión total")
ax.set_title("Desviación estándar por plataforma")

fig.savefig("sports_weights.png")

plt.show()
```
