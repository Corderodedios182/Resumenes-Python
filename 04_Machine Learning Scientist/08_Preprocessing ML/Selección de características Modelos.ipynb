### Seleccionando características de un vector de texto, para poder entrenar un algoritmo de clasificación Naive-Bayes.

*¿De que nos sirve ponerar con tfidf?* : Colocarle un valor númerico a las palabras nos permitirá usar esos valores para entrenar un modelo de clasificación. Como Naive-Bayes.

De la base volunteer veamos la columna title podemos observar que cada fila contiene una frase.

Ponderemos cada palabra de nuestra columna title.

Podemos tomar el 20% superior de las palabras ponderadas en todo el vector.

```python
volunteer["title"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#Vectorizando nuestro texto.
title_text = volunteer["title"]
tfidf_vec = TfidfVectorizer()
text_tfidf = tfidf_vec.fit_transform(title_text)

#Palabras organizadas con mayor a menor tf-idf.
pd.DataFrame(tfidf_vec.vocabulary_.keys(), tfidf_vec.vocabulary_.values()).reset_index().sort_values("index")
```

Esto nos indica el peso de cada palabra comparada contra todas las filas pero solo de la fila 0.

```python
#tf-idf primera frase en la primera fila y sus valores tf-idf.
print(text_tfidf[0])
```

```python
vocab = {v: k for k,v in tfidf_vec.vocabulary_.items()}

dict(zip(text_tfidf[0].indices, text_tfidf[0].data))
```

```python
def return_weights(vocab, vector, vector_index): 
    
    zipped = dict(zip(vector[vector_index].indices, vector[vector_index].data))
    
    return {vocab[i]: zipped[i] for i in vector[vector_index].indices}
```

**Palabras y peso en la frase de la primera fila.**

```python
print(return_weights(vocab, text_tfidf, 0))
```

En esta primera parte de la exploración de vectores de texto, vamos a agregar a esa función que aprendimos.

Devolveremos una lista de números con características.

En el siguiente ejercicio, escribiremos otra función para recopilar las palabras principales en todos los documentos, extraerlas y luego usar esa lista para filtrar nuestro text_tfidfvector.

```python
def return_weights(vocab, original_vocab, vector, vector_index, top_n):
    zipped = dict(zip(vector[vector_index].indices, vector[vector_index].data))
    
    zipped_series = pd.Series({vocab[i]:zipped[i] for i in vector[vector_index].indices})
    
    # Ordenando para sacar las n palabras más relevantes.
    zipped_index = zipped_series.sort_values(ascending=False)[:top_n].index
    return [original_vocab[i] for i in zipped_index]

# Palabras poneradas
print(return_weights(vocab, tfidf_vec.vocabulary_, text_tfidf, 0, 3))
```

Vamos a extraer las palabras principales de cada documento en el vector de texto, devolver una lista de índices de palabras y usar esa lista para filtrar el vector de texto a esas palabras principales.

```python
def words_to_filter(vocab, original_vocab, vector, top_n):
    filter_list = []
    for i in range(0, vector.shape[0]):
    
        filtered = return_weights(vocab, original_vocab, vector, i, top_n)
        filter_list.extend(filtered)
    
    return set(filter_list)

filtered_words = words_to_filter(vocab, tfidf_vec.vocabulary_, text_tfidf, 3)

filtered_text = text_tfidf[:, list(filtered_words)]

print(filtered_text)
```

**Modelo de clasificación de texto Naive Bayes, con nuestras opciones de selección del ejercicio anterior, en el volunteer conjunto de datos title y las category_desc columnas.**

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

y = volunteer["category_desc"]

X_train, X_test, y_train, y_test = train_test_split(filtered_text.toarray(), y, stratify=y)

nb = GaussianNB()
nb.fit(X_train, y_train)
print(nb.score(X_test, y_test))
```

### Reducción de Dimensionalidad

Una forma menos manual de reducir el tamaño de nuestro conjunto de datos es através de reducción de dimensionalidad.

Se puede realizar de forma líneal o no líneal.

PCA : Utiliza una transformación líneal para proyectar características en un espacio donde no están correlacionadas en absoluto.

Si bien las características se reducen, la varianza se captura de manera significativa al combinar características.

PCA captura en cada componente, la mayor cantidad posible de variación en el conjunto de datos. 

Nos facilita la toma de descición para la eliminación de componentes.

```python
wine.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA()
wine_X = wine.drop("Type", axis=1)

transformed_X = pca.fit_transform(wine_X)

#Impresión de la relación de varianza explicada,
#Revisar que componentes explican una mayor varianza en los datos.
print(pca.explained_variance_ratio_)
```

**Ahora que hemos ejecutado PCA en el wineconjunto de datos, intentemos entrenar un modelo con él.**

```python
from sklearn.neighbors import KNeighborsClassifier 

y = wine["Type"]

X_wine_train, X_wine_test, y_wine_train, y_wine_test = train_test_split(transformed_X, y)

knn = KNeighborsClassifier(n_neighbors=6)

knn.fit(X_wine_train, y_wine_train)

print(knn.score(X_wine_test, y_wine_test))
```
