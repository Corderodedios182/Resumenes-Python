```python
import pandas as pd

gapminder = pd.read_csv('gapminder.csv')

gapminder.head()
```

```python
Spain = gapminder[gapminder.country == 'Spain']
Mexico = gapminder[gapminder.country == 'Mexico']
Nepal = gapminder[gapminder.country == 'Nepal']
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot') #Habilita los estilos de R ggplot
#plt.style.use('tableau-colorblind10')

#Graficas independientes
plt.axes([0.05,0.05,0.425,0.9]) #Posicin 
plt.plot(Spain.year, Spain.lifeExp , color = 'red')
plt.xlabel('aos')
plt.ylabel('Esperanza de Vida')
plt.xlim(1960,2000)
plt.title('Espaa')

plt.axes([0.525,0.05,0.425,0.9])
plt.plot(Mexico.year, Mexico.lifeExp , color = 'pink')
plt.title('Mexico')

plt.axes([1.01,0.05,0.425,0.9])
plt.plot(Nepal.year, Nepal.lifeExp , color = 'green')
plt.title('Nepal')

plt.show()
```

```python
#Varias graficas distintas
#plt.subplot() genera una matriz 2x2 lugar 1,2,3
plt.subplot(2,2,1)
plt.plot(Spain.year, Spain.lifeExp , color = 'red')
plt.title('Espaa')

plt.subplot(2,2,2)
plt.plot(Mexico.year, Mexico.lifeExp , color = 'pink')
plt.title('Mexico')
cs_prom = int(Mexico.lifeExp.mean())
yr_prom = Mexico.year.mean()
plt.annotate('Promedio : ' + str(cs_prom), xy=(yr_prom, cs_prom), xytext=(yr_prom+10, cs_prom+10), arrowprops=dict(facecolor='black'))

plt.subplot(2,2,3)
plt.plot(Nepal.year, Nepal.lifeExp , color = 'green')
plt.title('Nepal')

plt.tight_layout()
plt.show()
```

```python
#Todas las graficas juntas
plt.plot(Spain.year, Spain.lifeExp , color = 'red', label = 'Espaa')
plt.plot(Mexico.year, Mexico.lifeExp , color = 'pink', label = 'Mexico')
plt.plot(Nepal.year, Nepal.lifeExp , color = 'green', label = 'Nepal')

plt.legend(loc='lower right')
plt.xlabel('aos')
plt.ylabel('Esperanza de Vida')
plt.ylim(20,90)
plt.title('Comparativo Esperanza de vida')

cs_min = Nepal.lifeExp.min().round()
yr_min = Nepal.year.min()

plt.annotate('Minimo : ' + str(cs_min), xy=(yr_min, cs_min), xytext=(yr_min+10, cs_min+10), arrowprops=dict(facecolor='black'))


plt.show()
```
